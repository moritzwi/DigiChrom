# Preprocessing: Spaltenauswahl für ML

```python
import pandas as pd
import numpy as np

df_raw = pd.read_csv(
    '/Users/moritzwinklmair/Library/CloudStorage/GoogleDrive-mowi9942@gmail.com/Meine Ablage/work/PhD/DigiChrom/DigiChrom_Pipeline/table_HARTCHROM_IPA_BC_Gold_Ontology.csv',
    sep=None, engine='python'
)

# --- Spalten klassifizieren ---
all_cols = df_raw.columns.tolist()

# Experiment-ID (enthält IPA-xxx / BC-xxx Prefix als Org-Identifier)
experiment_id_col = '01_preparationsubstrate_objectraw_objectid_01f8s0n1'

# Zu droppende Kategorien
drop_unit      = [c for c in all_cols if '_unitsymbol_' in c]
drop_ids       = [c for c in all_cols if any(x in c for x in [
                    '_processid_', '_electrolyteid_', '_electrolytesampleid_',
                    '_analysisid_', '_objectid_', '_electrolytesampleid_'
                  ]) and c != experiment_id_col]
drop_procedure = [c for c in all_cols if '_procedure__sp__name_string_' in c
                                      or '_organization__sp__name_string_' in c]
drop_dates     = [c for c in all_cols if 'yyyy__dash__mm' in c]
drop_meta      = [c for c in all_cols if any(x in c for x in [
                    '_suppliername_', '_batchnumber_', '_articlenumber_',
                    '_file__sp__name_', '_equipment__sp__name_',
                    '_productname_', '_consumablename_', '_operator__sp__name_',
                    '_prename__sp__'
                  ])]

# __calc-Spalten: bereinigt/vereinheitlicht → Originale droppen
calc_cols = [c for c in all_cols if c.endswith('__calc')]

# Finde das nicht-calc Gegenstück anhand des Parameternamens
# z.B. "pickling_treatmenttime_value_00br3p9f__calc" → suche "pickling_treatmenttime_value_00br3p9f"
drop_noncalc = []
for calc_col in calc_cols:
    base = calc_col.removesuffix('__calc')
    if base in all_cols:
        drop_noncalc.append(base)

# Alle zu droppenden Spalten
to_drop = set(drop_unit + drop_ids + drop_procedure + drop_dates + drop_meta + drop_noncalc)

# Sicherstellen dass experiment_id_col nicht gedroppt wird
to_drop.discard(experiment_id_col)

df = df_raw.drop(columns=[c for c in to_drop if c in df_raw.columns])

# Komplett leere Spalten droppen
df = df.dropna(axis=1, how='all')

# __calc Suffix aus Spaltennamen entfernen (Lesbarkeit)
df = df.rename(columns={c: c.removesuffix('__calc') for c in df.columns if c.endswith('__calc')})

# --- Überblick ---
value_cols   = [c for c in df.columns if '_value_' in c]
cat_cols     = [c for c in df.columns if 'currentform' in c or 'coverage' in c or 'medium' in c]
remain_other = [c for c in df.columns if c not in value_cols + cat_cols + [experiment_id_col]]

print(f"Zeilen:               {len(df)}")
print(f"Spalten gesamt:       {len(df.columns)}  (vorher: {len(df_raw.columns)})")
print(f"  Experiment-ID:      1")
print(f"  Numerische Values:  {len(value_cols)}")
print(f"  Kategorische:       {len(cat_cols)}")
print(f"  Sonstige:           {len(remain_other)}")
if remain_other:
    print(f"    -> {remain_other}")
print()

# Fill-Rate der Value-Spalten
print("Fill-Rate Value-Spalten:")
fills = df[value_cols].notna().mean().sort_values(ascending=False)
for col, rate in fills.items():
    short = '_'.join(col.split('_')[1:-1])
    print(f"  {rate:5.1%}  {short[:70]}")
```
